```python
import pandas as pd
```

```python
DownloadedPath = "Resources/JC-20"
pathEnd ="-citibike-tripdata.csv"
```

```python
df = pd.read_csv(DownloadedPath+"2101"+pathEnd)
df.to_csv("ChallengeData/Dataset.csv",index=False, header=True)
df.head()
```

```python
myYear="20"
for month in range(6,13):
    if month<10:
        myMonth="0"+str(month)
    else:
        myMonth=str(month)
    mydf=pd.read_csv(DownloadedPath+myYear+myMonth+pathEnd,encoding='utf-8')
    mydf.to_csv("ChallengeData/Dataset.csv",mode="a", index=False, header=False)
```
